```python
import faker
from faker import Faker

fake = Faker()

# Specify the number of records you want to generate
num_records = 10000

# Create a list to store the fake procurement data
procurement_data = []

for i in range(num_records):
    record = {
        "spend": fake.pydecimal(left_digits=5, right_digits=2, positive=True),
        "date": fake.date_between(start_date="-4y", end_date="today"),
        "supplier": fake.random_element(["A", "B", "C","D","E"]),
        "region": fake.country(),
        "relatedBOM_name": fake.word(),  
        "quantity": fake.random_int(min=1, max=100),  
        "spend_type": fake.random_element(["Direct", "Indirect", "Capital"]),  
        "franchise": fake.word(),  
        "material_category": fake.word(),  
        "subcategory": fake.word(),  
        "warehouse": fake.address().replace("\n", " "),  
    }
    procurement_data.append(record)


print(procurement_data)


```

```python
# Save the dataset to a CSV file
import csv

with open("procurement_data.csv", "w", newline="") as csvfile:
    fieldnames = [
        "spend",
        "date",
        "supplier",
        "region",
        "relatedBOM_name",
        "quantity",
        "spend_type",
        "franchise",
        "material_category",
        "subcategory",
        "warehouse",
    ]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(procurement_data)
```

```python
import pandas as pd

procurement_df = pd.DataFrame(procurement_data)
procurement_df.head()
```
